# 311 Data

```python
df311_og = pd.read_csv('311_Service_Requests_from_2010_to_Present_final.csv')
df311_og.tail()
```

```python
df311 = df311_og[['Created Date','Borough','Complaint Type','Descriptor','Location Type','Incident Zip','Community Board', 'Location']]
```

#### Clearning Community District Column

```python
df311['Borough'].replace({'BROOKLYN':'BK',"QUEENS":'QN','BRONX':'BX','MANHATTAN':'MN','STATEN ISLAND':'SI'},inplace=True)
df311['CD'] = df311.Borough + df311['Community Board'].str[0:2]
df311.drop(['Borough','Community Board'],axis=1,inplace=True)
df311.head()
```

#### Temporal Dimension

```python
import datetime
```

```python
df311['Created Date']= pd.to_datetime(df311['Created Date'])
```

```python
df311['Month'] = df311['Created Date'].dt.month
df311['Year'] = df311['Created Date'].dt.year
df311['Year_Month'] = pd.to_datetime(df311['Created Date']).dt.to_period('M')
df311.head()
```

# Rating Data 

```python
rating_og = pd.read_csv('Scorecard_Ratings_all.csv')
rating_og.head()
```

```python
rating = rating_og.drop(['Borough','Community Board','Cleaning Section'],axis=1)
```

### Temporal Dimension

```python
rating['Month']=pd.to_datetime(rating['Month'], format='%Y / %m')
rating['Year_Month'] = pd.to_datetime(rating['Month']).dt.to_period('M')
rating.head()
```

```python
rating.dtypes
```

```python
rating = rating.sort_values(by='Year_Month')
print(rating['Year_Month'].unique())
```

```python
rating['Year_Month']
```

```python
rating = rating[(rating['Year_Month'] >= '2016-11')]
rating.reset_index(drop=True, inplace=True)
rating.head()
```

```python
rating.size
```

```python
len(rating['Acceptable Streets %'].isna())
```

```python
rating.dtypes
```

```python
rating['Month'] = rating['Year_Month'].dt.month
rating['Year'] = rating['Year_Month'].dt.year
rating.head()
```

```python
rating['Acceptable Streets %'].isna().sum()
```

```python
rating['Acceptable Sidewalks %'].isna().sum()
```

```python
rating['Acceptable Streets % - Previous Month'].isna().sum()
```

```python
rating['Acceptable Streets %'] = np.where(rating['Acceptable Streets %'].isnull(), rating['Acceptable Streets % - Previous Month'], rating['Acceptable Streets %'])
```

```python
rating['Acceptable Streets %'] = np.where(rating['Acceptable Streets %'].isnull(), rating['Acceptable Streets % - Previous Year'], rating['Acceptable Streets %'])
```

```python
rating['Acceptable Streets %'] = np.where(rating['Acceptable Streets %'].isnull(), rating['Acceptable Streets % - Previous Fiscal Quarter'], rating['Acceptable Streets %'])
```

```python
rating['Acceptable Streets %'].isna().sum()
```

```python
rating['Acceptable Sidewalks %'] = np.where(rating['Acceptable Sidewalks %'].isnull(), rating['Acceptable Sidewalks % - Previous Month'], rating['Acceptable Sidewalks %'])
```

```python
rating['Acceptable Sidewalks %'] = np.where(rating['Acceptable Sidewalks %'].isnull(), rating['Acceptable Sidewalks % - Previous Year'], rating['Acceptable Sidewalks %'])
```

```python
rating['Acceptable Sidewalks %'] = np.where(rating['Acceptable Sidewalks %'].isnull(), rating['Acceptable Sidewalks % - Previous Fiscal Quarter'], rating['Acceptable Sidewalks %'])
```

```python
rating['Acceptable Sidewalks %'].isna().sum()
```

```python
rating['Acceptable Streets %'] = rating['Acceptable Streets %'].fillna(100)
```

```python
rating['Acceptable Streets %'].isna().sum()
```

```python
rating['Acceptable Sidewalks %'] = rating['Acceptable Streets %'].fillna(100)
```
